```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # Correct URL for cities with more than two words in their name (e.g. New York City)
    city = city.replace(" ", "%20")
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Generating URL
base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "units=Imperial"

url = base_url + units + "&APPID=" + api_key
```

```python
# Empty Lists for Appending
name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

# Call Counter
counter = 1

# Initial Print Statement
p = ("Beginning Data Retrieval\n"
    "-------------------------------")
print(p)

for city in cities:
    try: 
        response = requests.get(f"{url}&q={city}").json() 
        name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Processing Record {counter} | {city_record}")
        print(f"{url}&q={city}")
        
        counter = counter + 1
        
        # So the API doesn't crash...
        time.sleep(0.5)
        
    except:
        print("City not found. Skipping...")
    continue
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
cityweather_dict = {
    "City": name,
    "Cloudiness":cloudiness, 
    "Country":country,
    "Date":date, 
    "Humidity": humidity,
    "Lat":lat, 
    "Lng":lng, 
    "Max Temp": max_temp,
    "Wind Speed":wind_speed
}

# Weather DataFrame
weatherpy_df = pd.DataFrame(cityweather_dict)

# Count of weather data values 
weatherpy_df.count()
```

```python
# Save data frame to CSV
weatherpy_df.to_csv('output_weatherpy.csv')

# Display data frame head 
weatherpy_df.head()
```
